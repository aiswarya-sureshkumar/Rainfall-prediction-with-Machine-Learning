# *  Handling the missing values



```python
# Filling the missing float values with mean
df['MinTemp']=df['MinTemp'].fillna(df['MinTemp'].mean())
df['MaxTemp']=df['MaxTemp'].fillna(df['MaxTemp'].mean())
df['Rainfall']=df['Rainfall'].fillna(df['Rainfall'].mean())
df['Evaporation']=df['Evaporation'].fillna(df['Evaporation'].mean())
df['Sunshine']=df['Sunshine'].fillna(df['Sunshine'].mean())
df['WindGustSpeed']=df['WindGustSpeed'].fillna(df['WindGustSpeed'].mean())
df['WindSpeed9am']=df['WindSpeed9am'].fillna(df['WindSpeed9am'].mean())
df['WindSpeed3pm']=df['WindSpeed3pm'].fillna(df['WindSpeed3pm'].mean())
df['Humidity9am']=df['Humidity9am'].fillna(df['Humidity9am'].mean())
df['Humidity3pm']=df['Humidity3pm'].fillna(df['Humidity3pm'].mean())
df['Pressure9am']=df['Pressure9am'].fillna(df['Pressure9am'].mean())
df['Pressure3pm']=df['Pressure3pm'].fillna(df['Pressure3pm'].mean())
df['Cloud9am']=df['Cloud9am'].fillna(df['Cloud9am'].mean())
df['Cloud3pm']=df['Cloud3pm'].fillna(df['Cloud3pm'].mean())
df['Temp9am']=df['Temp9am'].fillna(df['Temp9am'].mean())
df['Temp3pm']=df['Temp3pm'].fillna(df['Temp3pm'].mean())
```

```python
# Filling the missing object values with mode
df['WindGustDir']=df['WindGustDir'].fillna(df['WindGustDir'].mode()[0])
df['WindDir9am']=df['WindDir9am'].fillna(df['WindDir9am'].mode()[0])
df['WindDir3pm']=df['WindDir3pm'].fillna(df['WindDir3pm'].mode()[0])
df['RainToday']=df['RainToday'].fillna(df['RainToday'].mode()[0])
```

```python
df.isna().sum()
```



# *   Identification and Removal of outliers - Interquartile range (IQR) method



```python
# Identification of outliers by plotting boxplot for numerical columns
num_columns=['MinTemp','MaxTemp','Rainfall','Evaporation','Sunshine','WindGustSpeed','WindSpeed9am','WindSpeed3pm','Humidity9am','Humidity3pm','Pressure9am','Pressure3pm','Cloud9am','Cloud3pm','Temp9am','Temp3pm','RISK_MM']
for i in df:
  if i in num_columns:
     plt.figure()
     plt.gca().set_title(i)
     df.boxplot([i])
```

```python
# Removing the outilers using IQR method
def remove_outliers(df):
  outlier_indices=[]
  columns=['MinTemp','MaxTemp','Rainfall','Evaporation','Sunshine','WindGustSpeed','WindSpeed9am','WindSpeed3pm','Humidity9am','Humidity3pm','Pressure9am','Pressure3pm','Cloud9am','Cloud3pm','Temp9am','Temp3pm','RISK_MM']
  for i in columns:
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)
    IQR=Q3-Q1
    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR
    outliers=df.index[(df[i]<lower_limit)|(df[i]>upper_limit)]
    outlier_indices.extend(outliers)
  df_cleaned=df.drop(outlier_indices)
  return df_cleaned
```

```python
df1=remove_outliers(df)
df1.shape
```



# *   Label encoding



```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df1['Date']=lb.fit_transform(df1['Date'])
df1['Location']=lb.fit_transform(df1['Location'])
df1['WindGustDir']=lb.fit_transform(df1['WindGustDir'])
df1['WindDir9am']=lb.fit_transform(df1['WindDir9am'])
df1['WindDir3pm']=lb.fit_transform(df1['WindDir3pm'])
df1['RainToday']=lb.fit_transform(df1['RainToday'])
```

```python
df1.dtypes
```

# **Splitting into Training and Testing data**

---



```python
# Separating the input and output features
x=df1.iloc[:,:-1]
y=df1.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

# **Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train_normalized=scaler.transform(x_train)
x_test_normalized=scaler.transform(x_test)
```

# **BALANCING**

```python
# Balancing by Resampling Technique
from imblearn.under_sampling import RandomUnderSampler
from sklearn.datasets import make_classification
under_sampler = RandomUnderSampler(random_state=42)
x_train_balanced,y_train_balanced = under_sampler.fit_resample(x_train_normalized,y_train)
```

```python
sns.countplot(x=y_train_balanced)
```

# **FEATURE SELECTION**

---



```python
# Chi-squared test
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
selector=SelectKBest(chi2, k=18)
x_train_selected=selector.fit_transform(x_train_balanced,y_train_balanced)
x_test_selected=selector.transform(x_test_normalized)

# To get the indices of the selected features
selected_feature_indices=selector.get_support(indices=True)

# To get the names of the selected features
selected_feature_names=x.columns[selected_feature_indices]

# To print the selected feature names
print("Selected Features:")
print(selected_feature_names)
```

# **Model creation and Performance evaluation**

---



```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
knn=KNeighborsClassifier(n_neighbors=7)
naive=GaussianNB()
vector=SVC()
decision=DecisionTreeClassifier(criterion='entropy')
random=RandomForestClassifier(n_estimators=100)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
lst_models=[knn,naive,vector,decision,random]
accuracy=[]
for i in lst_models:
  print(i)
  i.fit(x_train_selected,y_train_balanced)
  y_pred=i.predict(x_test_selected)
  score=accuracy_score(y_test,y_pred)
  accuracy.append(score)
  print("******************************************************************")
  print(confusion_matrix(y_test,y_pred))
  print("******************************************************************")
  print(accuracy_score(y_test,y_pred))
  print("******************************************************************")
  print(classification_report(y_test,y_pred))
```

```python
models=['KNN','Naive bayes','SVM','Decision tree ','  Random forest']
plt.bar(models,accuracy,color=['red','green','yellow','purple','blue'])
plt.xlabel('Models',color='red')
plt.ylabel('Accuracy scores',color='red')
plt.title('Models vs Accuracy score')
plt.show()
```

```python
# Error plot for Decision Tree model
max_depth_values = range(1, 21)
train_errors = []
test_errors = []

for max_depth in max_depth_values:
    decision = DecisionTreeClassifier(max_depth=max_depth, random_state=42)
    decision.fit(x_train_selected, y_train_balanced)
    y_train_pred = decision.predict(x_train_selected)
    y_test_pred = decision.predict(x_test_selected)

    train_accuracy = accuracy_score(y_train_balanced, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)

    train_errors.append(1 - train_accuracy)
    test_errors.append(1 - test_accuracy)

plt.plot(max_depth_values, train_errors, label='Training Error', marker='o')
plt.plot(max_depth_values, test_errors, label='Testing Error', marker='o')
plt.xlabel('Max Depth of the Decision Tree')
plt.ylabel('Error Rate')
plt.title('Decision Tree Classifier Error Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Error plot for Random forest model

n_estimators_range = [10, 50, 100, 150, 200]
train_errors = []
test_errors = []

for n_estimators in n_estimators_range:
    random = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
    random.fit(x_train_selected, y_train_balanced)
    y_train_pred = random.predict(x_train_selected)
    y_test_pred = random.predict(x_test_selected)

    train_accuracy = accuracy_score(y_train_balanced, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)

    train_errors.append(1 - train_accuracy)
    test_errors.append(1 - test_accuracy)

plt.plot(n_estimators_range, train_errors, label='Training Error')
plt.plot(n_estimators_range, test_errors, label='Testing Error')
plt.xlabel('Number of Estimators')
plt.ylabel('Error Rate')
plt.title('Random Forest Classifier Error Plot')
plt.legend()
plt.show()
```
